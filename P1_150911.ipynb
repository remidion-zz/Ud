###Background Information

In a Stroop task, participants are presented with a list of words, with each word displayed in a color of ink. The participantâ€™s task is to say out loud the color of the ink in which the word is printed. ***The task has two conditions: a congruent words condition, and an incongruent words condition. ***

*In the **congruent** words condition, the words being displayed are color words whose names match the colors in which they are printed: for example RED, BLUE.* 

*In the **incongruent** words condition, the words displayed are color words whose names do not match the colors in which they are printed: for example PURPLE, ORANGE. In each case, we measure the time it takes to name the ink colors in equally-sized lists. Each participant will go through and record a time from each condition.*

## Questions
###1. What is our independent variable? 
A dependant variable is defined as "The independent variable of a study is the variable that experimenters choose to manipulate; it is usually plotted along the x-axis of a graph" ([Reference 1] [1])

According to this definition, the independant variable is the **condition of the word**, either congruent or incongruent.


###What is our dependent variable?
An independant variable is defined as "The dependent variable of a study is the variable that experimenters choose to measure during an experiment; it is usually plotted along the y-axis of a graph" ([Reference 1] [1])

Again, according to this definition, the **time taken to execute the task (verbally identifying the color of the ink)** is the dependant variable.

###2. What is an appropriate set of hypotheses for this task? What kind of statistical test do you expect to perform? Justify your choices.
The first question to ask is weither this phenomenon corresponds to a one that would be normally distributed.  Many useful statistical tests only apply to *normally distributed samples*.  
The hypothesis must correlate the stroop effect to the time to execute the task. According to my own experience doing the experiement, I would postulate that it takes more time __on average__ to verbally identify the color on the incongruent task.  
Therefore, the null hypothesis is that the stroop effect does not affect the mean time  ($H_0 : \mu_I-\mu_c \leq 0$).
The alternative hypothesis is that the stroop effect *does*, in fact, increases the mean time ($H_a : \mu_i > \mu_c$).

Given that we don't know the population's parameters, we will use the sample's characteristic's to determine the meaningfulness of the difference of the two means by applying the [paired t-test][4] and calculating Cohen's d.  

The paired t-test is most appropriate since we are looking for a difference between the means of two dependent observations (same participants in both measures) and want to test if the mean differs significantly from 0. The z-test has been ruled out because we use sample and have no parameter for the population and the sample size is less than 30.

###5. Now, perform the statistical test and report your results. What is your confidence level and your critical statistic value? Do you reject the null hypothesis or fail to reject it? Come to a conclusion in terms of the experiment task. Did the results match up with your expectations?

####Hypothesis testing
We can apply the t-test ([Reference 1][1]) in order to examine if the sample fall within the confidence interval.  If it does, then the null hypothesis would be proved right.  Inversely, if it doesn't, then the alternative hypothesis will be supported by the data.  We know that the sample size is n=24 (thus our degrees of freedom (df) value is 23).

We will consider a confidence level of 95% which sets the critical statistic value ($t_{crit}$) for a one-tail test to 1.714.

We want to see if the incongruent dataset presents significantly different values than the congruent dataset.   

\begin{align}
t = \frac{\bar{x_d}}{\frac{s_d}{\sqrt{n}}} \\
t = \frac{7.965}{\frac{4.865}{\sqrt{24}}} \\
t=  8.021\\
\end{align}

This value is way beyond the $t_{crit}$ value, it corresponds to a p-value of 0.0001 ([Reference 4][5]) and the difference is extremely statistically significant and that we must reject the null hypothesis.  


We can also use Cohen's d measure to compare the two datasets as if we were unaware of the experiment protocol (blind study) and assume they are two independant datasets of the same population.  

The Cohen's d measures the distance between two means in terms of a unique and common standard deviations. It can be found using the following formula:

\begin{align}
d = \frac{\bar{x_i}-\bar{x_c}}{\sqrt{s_i^2+s_c^2}} \\
d = \frac{22.016-14.051}{\sqrt{4.797^2+3.559^2}} \\
d=  \frac{7.965}{5.973}=1.333 \\
\end{align}

This measure shows that the two data sets differ on average by more than a full standard deviation of each other, which is statistically significant and indicates major differences between the two sample, as expected. 

The result match the expectations and promotes the idea that the brain performs less efficiently when interpreting two opposing inputs simultaneously.
